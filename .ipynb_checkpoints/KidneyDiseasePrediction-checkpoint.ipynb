## Predicting   chronic kidney disease 

**Goal of this notebook is to help physician to diagnosis the patients with chronic disease basis of some lab work**

**what is chronic kidney disease?** 
Chronic kidney disease (CKD) means your kidneys are damaged and can’t filter blood the way they should. The disease is called “chronic” because the damage to your kidneys happens slowly over a long period of time. *U.S. Department of Health and Human Services*


**Who is more likely to develop CKD?**

**Diabetes**. **Diabetes is the leading cause of CKD. High blood glucose**, also called blood sugar, from diabetes can damage the blood vessels in your kidneys. Almost 1 in 3 people with diabetes has CKD

**High blood pressure**. **High blood pressure is the second leading cause of CKD**. Like high blood glucose, high blood pressure also can damage the blood vessels in your kidneys. **Almost 1 in 5 adults with high blood pressure has CKD**.

**Heart disease**. Research shows a link between kidney disease and heart disease. People with heart disease are at **higher risk for kidney disease, and people with kidney disease are at higher risk for heart disease**. 

**Family history of kidney failure**. If your mother, father, sister, or brother has kidney failure, you are at risk for CKD. Kidney disease tends to run in families.
*U.S. Department of Health and Human Services*


**Being African-American, Hispanic Native American or Asian** *American Kidney Fund*

**Being over 60 years old** *American Kidney Fund*

## Lets get to know more about features

There are 25 features but it looks like some features have more or less the same info like blood pressure/hypertension? anemia and red blood count/red blodd cell/anemia!

**Albumin:**Albuminuria is a sign of kidney disease and means that you have too much albumin in your urine. Albumin is a protein found in the blood. A healthy kidney doesn’t let albumin pass from the blood into the urine. A damaged kidney lets some albumin pass into the urine. The less albumin in your urine, the better.

**Hypertension/High Blood Pressure**
 
The key difference between hypertension and high blood pressure is that the hypertension is a medical diagnosis where blood pressure is persistently elevated at or above 140/90 mm Hg for most adults.  Whereas, high blood pressure refers to a nonspecific elevation of the blood pressure above 130/80 mmHg. Hypertension seems to be the one matters more for ckd!


**A random blood glucose/sugar/diabetes mellitus** random blood glucose measures how much sugar (glucose) you have in your blood at the time you are tested. This test is sometimes called the casual glucose test. If your blood glucose level is 200 mg/dL or higher, you likely have diabetes mellitus.


**appetite** Many Chronic Kidney Disease patients, particularly those on dialysis, have problems with their appetite. Some complain that they do not want to eat because they always "feel full," or that they have "a strange taste in their mouths" or "no taste for food at all.

**urine-specific gravity** In chronic kidney disease, the urine-specific gravity is neither concentrated nor dilute; it falls in a middle range known as isothenuric because the impaired kidneys no longer have the ability to concentrate or dilute the urine.

**anemia/red blood cell/red blood cell/Hemoglobin** An individual suffering from chronic kidney disease is most likely to suffer from anemia as well.
Anemia caused by **decreased red blood cell** production. 

Red blood cells are a very essential part of our blood and as a matter of fact, they are the most plentiful type of blood cells in our body. Hemoglobin or Hgb is a protein that is found inside a red blood cell and is responsible for carrying the oxygen coming from the lungs to the different parts of the body.



**pedal edema(Swollen feet) ** is the accumulation of fluid in the feet and lower legs. When you have kidney disease, extra fluid and sodium in your circulation may cause edema.

**bacteria/White blood cells/pus cell** People with kidney disease  can be more prone to infection because of related conditions such as diabetes, inadequate calorie and protein intake. they are are sign of infection.**Pus cells** are a mixture of dead tissue, bacteria, and white blood cells

**coronary artery disease**Chronic kidney disease (CKD) is an independent risk factor for coronary artery disease (CAD). Coronary artery disease is the leading cause of morbidity and mortality in patients with CKD



**blodd urea** If increased amounts of urea are produced by the liver or if the kidneys are not working properly and have difficulty filtering wastes out of the blood, then urea concentrations will rise in the blood. If significant liver damage or disease inhibits the production of urea, then BUN concentrations may fall.

**serum creatinine** Creatinine has been found to be **a fairly reliable indicator of kidney function**. Elevated creatinine level signifies impaired kidney function or kidney disease. As the kidneys become impaired for any reason, the creatinine level in the blood will rise due to poor clearance of creatinine by the kidneys.

**sodium** When your sodium levels are low, your kidneys essentially hold on to the sodium . When sodium levels are high, your kidneys excrete the excess in urine. But if for some reason your kidneys can't eliminate enough sodium, the sodium starts to accumulate in your blood


**packed cell volume** lower PCV worsen the severity of ckd

# How to deal with missing values?
**There is a high correlation between missingness of features** For example correlation between missingness of white blood cell and red blood cell is 0.8!  that means it would,nt be a good idea to use correlated features to fill their  missing values.

I dont want to remove the rows with missing values either. So i fill the missing values with median.

```python
# filling  missing values of numerical attributes with the median. 
for atr_n in attr_num:
    df[atr_n].fillna(value=df[atr_n].median(),inplace=True)
    #filling missing values of categorical attributes with most frequent values
for atr_c in attr_cat:
    df[atr_c].fillna(value=df[atr_c].mode(),inplace=True)
```

```python
# converting  categorical variable into dummy  variables.

df=pd.get_dummies(df,drop_first=True,columns=attr_cat)
```

# BaseLine Model

I am going to use Logistic regression as my basline model

```python
LogR=LogisticRegression(solver="liblinear",random_state=42)
```

```python

accuracy = cross_val_score(LogR, train_set, train_target, cv=3, scoring='accuracy')
Mean=100*accuracy.mean()
SD =100*accuracy.std()
print( "  Mean Accuracy: ", Mean, "   Standard Deviation: ", SD)
```

The accuracy is very high! But I am not very excited about it since i have an imbalance positive and negative samples( ~67% ckd class and and  33%   Not_ckd_class)! the way accuracy works, finds the total number of correct predictions! both true positive and true negative prediction! but in this case since we have way more ckd class, accuracy is not very reliable! let’s look at a exaggerated example. If not_ckd class was 90% of target, then if we have model the labels all target as not_ckd class,  the accuracy will be 90%!!! since 90% of targets are not_ckd and been predicted correctly! so accuracy is not a good evaluation method for this imbalance target class!

In this model both precision and recall are important! so I use f1_score! f1_score will be low if any of scores are low! in the other word, it will be high if both precision and recall are high! since positive and negative samples are imbalanced I prefer to use f1_score than roc_auc score since ROC averages over all possible thresholds!

```python

target_train_pred = cross_val_predict(LogR, train_set, train_target, cv=3)
F1_score=f1_score(train_target, target_train_pred)
print( 'F1_score:',F1_score )
```

```python
F1_score=[];
for k in np.arange(2,11):
    target_train_pred = cross_val_predict(LogR, train_set, train_target, cv=k)
    tmp=f1_score(train_target, target_train_pred)
    F1_score.append(tmp)
    #print( 'F1_score:',F1_score,'k fold =',k ) 
fig, axes=plt.subplots(figsize=(10,5))
axes.plot(F1_score)
plt.xlabel('Number of Folds') 
plt.ylabel('F1 Score')
```

Let's stick with k=3 then!

 For the next step I am going to check if the model is overfitting the data! Let's look at learning curve for it! The learning curve will show the performance of the model as the samples grow!

```python
def Plot_Learning_Curve(Estimator,Bins,Folds,train_set,targe_tset):
    from sklearn.model_selection import learning_curve
    train_sizes, train_scores, validation_scores = learning_curve(Estimator,X=train_set, y=train_target,train_sizes=np.linspace(0.1, 1.0, Bins),cv=Folds,scoring = 'f1',shuffle=True)
    train_scores_mean = train_scores.mean(axis = 1)
    validation_scores_mean = validation_scores.mean(axis =1 )
    plt.plot(train_sizes, train_scores_mean,'r-+', label = 'Train data error')
    plt.plot(train_sizes, validation_scores_mean, 'b-',label = 'Validation data error')
    plt.ylabel('f1_score', fontsize = 10)
    plt.xlabel('Train  dataset size', fontsize = 10)
    plt.legend(framealpha=1, frameon=True);
```

```python
Plot_Learning_Curve(RandFor,30,3,train_set,train_target)
```

## Pre Prunning using grid search

```python
n_estimators=np.arange(10,200,10)
max_features = ['auto', 'sqrt']
max_depth=[10,20,50,100,300]
min_samples_split = [2,4,6,20]
min_samples_leaf = [1, 2, 4,10,30,50]
bootstrap = [True, False]
grid_search = {'n_estimators': n_estimators,
               'max_features': max_features,'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,  'bootstrap': bootstrap}
                           
```

```python
RNF=RandomForestClassifier()
param_search = RandomizedSearchCV(estimator = RNF, param_distributions = grid_search, n_iter = 100, cv = 4, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
param_search.fit(train_set, train_target)
param_search.best_params_
```
